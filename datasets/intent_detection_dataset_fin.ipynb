```python
user_profile_data = load_dataset("Akshaydevaraboina/workout_dataset") # colname Context
#more_user_profile_data = load_dataset("CristiD7/Comprehensive_7Day_Workout_Plans_100")
ask_advice_data = load_dataset("PandurangMopgar/fitness_data") # colname question, filter by answer
#more_advice_data = load_dataset("PandurangMopgar/fitness_data")
```

```python
ask_advice_data = ask_advice_data['train'].to_pandas()
```

```python
user_profile_data = user_profile_data['train'].to_pandas()
```

```python
import pandas as pd
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
from google.colab import files
uploaded = files.upload()
with open("fitness_diet_diary_1000.txt", 'r', encoding="utf-8") as f:
    text=f.read()

sentences = sent_tokenize(text)
store_log_data = pd.DataFrame(sentences, columns=["input"])
```

```python
# extract just the relevant column from each dataset
# add labels
# reformat headers to match
```

```python
user_profile_data = user_profile_data[["Context"]]
user_profile_data["label"] = "user_profile"
user_profile_data = user_profile_data.rename(columns={"Context": "input"})
```

```python
ask_advice_data['label'] = "ask_advice"
ask_advice_data = ask_advice_data[["instruction", "label"]]
ask_advice_data = ask_advice_data[ask_advice_data['instruction'] != 'instruction']
ask_advice_data = ask_advice_data.rename(columns={"instruction": "input"})
```

```python
store_log_data['label'] = "store_log"
```

```python
print(ask_advice_data)
```

```python
# merge all 3 datasets
intent_data = pd.concat([user_profile_data, store_log_data, ask_advice_data], ignore_index=True)
```

```python
#shuffle rows
intent_data = intent_data.sample(frac=1).reset_index(drop=True)
```

```python
intent_data = intent_data.applymap(lambda x: x.replace('Can you create a 7-day workout plan for me?', '') if isinstance(x, str) else x)
```

```python
intent_data.to_csv("intent_data_ad_log_pro.csv", index=False)
```
